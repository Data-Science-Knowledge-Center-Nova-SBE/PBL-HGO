```python
import pandas as pd
```

```python
df = pd.read_csv("downloads/data_with_meds.csv", sep=",")
df = df.dropna(subset=['Texto'])
df
```

```python
import matplotlib.pyplot as plt

plt.hist(df['medication_level_1'], bins=10, alpha=0.25, label='medication_level_1')
plt.hist(df['medication_level_2'], bins=10, alpha=0.25, label='medication_level_2')
plt.hist(df['medication_level_3'], bins=10, alpha=0.25, label='medication_level_3')

plt.axvline(df['medication_level_1'].mean(), color='blue', linestyle='dashed', linewidth=1)
plt.axvline(df['medication_level_2'].mean(), color='red', linestyle='dashed', linewidth=1)
plt.axvline(df['medication_level_3'].mean(), color='green', linestyle='dashed', linewidth=1)


plt.text(df['medication_level_1'].mean() + 1, 450, 'Mean: {:.2f}'.format(df['medication_level_1'].mean()), color='blue')
plt.text(df['medication_level_2'].mean() + 1, 400, 'Mean: {:.2f}'.format(df['medication_level_2'].mean()), color='red')
plt.text(df['medication_level_3'].mean() + 1, 350, 'Mean: {:.2f}'.format(df['medication_level_3'].mean()), color='green')


plt.legend(loc='upper right')
plt.show()
```

```python
plt.hist(df['medication_level_1'], bins=10, alpha=0.25, label='medication_level_1', density= True)
plt.hist(df['medication_level_2'], bins=10, alpha=0.25, label='medication_level_2', density= True)
plt.hist(df['medication_level_3'], bins=10, alpha=0.25, label='medication_level_3', density= True)


plt.legend(loc='upper right')
plt.show()
```

```python
import seaborn as sns


data_filtered_med1 = df.query('(medication_level_1 != 0)')
data_filtered_med2 = df.query('(medication_level_2 != 0)')
data_filtered_med3 = df.query('(medication_level_3 != 0)')

plt.hist(data_filtered_med1['medication_level_1'], bins=10, alpha=0.25, label='medication_level_1')
plt.hist(data_filtered_med2['medication_level_2'], bins=10, alpha=0.25, label='medication_level_2')
plt.hist(data_filtered_med3['medication_level_3'], bins=10, alpha=0.25, label='medication_level_3')

plt.axvline(data_filtered_med1['medication_level_1'].mean(), color='blue', linestyle='dashed', linewidth=1)
plt.axvline(data_filtered_med2['medication_level_2'].mean(), color='red', linestyle='dashed', linewidth=1)
plt.axvline(data_filtered_med3['medication_level_3'].mean(), color='green', linestyle='dashed', linewidth=1)

plt.text(data_filtered_med1['medication_level_1'].mean() + 1, 450, 'Mean: {:.2f}'.format(data_filtered_med1['medication_level_1'].mean()), color='blue')
plt.text(data_filtered_med2['medication_level_2'].mean() + 1, 400, 'Mean: {:.2f}'.format(data_filtered_med2['medication_level_2'].mean()), color='red')
plt.text(data_filtered_med3['medication_level_3'].mean() + 1, 350, 'Mean: {:.2f}'.format(data_filtered_med3['medication_level_3'].mean()), color='green')


plt.legend(loc='upper right')
plt.show()
```

## Symptoms Stats

```python
df = pd.read_csv("downloads/data_with_symptoms.csv", sep=",")
df = df.dropna(subset=['Texto'])
df.head()
```

```python
plt.hist(df['symptom_1'], bins=10, alpha=0.25, label='symptom_1')
plt.hist(df['symptom_0'], bins=10, alpha=0.25, label='symptom_0')

plt.axvline(df['symptom_1'].mean(), color='blue', linestyle='dashed', linewidth=1)
plt.axvline(df['symptom_0'].mean(), color='red', linestyle='dashed', linewidth=1)


plt.text(df['symptom_1'].mean() + 1, 450, 'Mean: {:.2f}'.format(df['symptom_1'].mean()), color='blue')
plt.text(df['symptom_0'].mean() + 1, 400, 'Mean: {:.2f}'.format(df['symptom_0'].mean()), color='red')


plt.legend(loc='upper right')
plt.show()
```

```python
plt.hist(df['symptom_1'], bins=10, alpha=0.25, label='symptom_1', density=True)
plt.hist(df['symptom_0'], bins=10, alpha=0.25, label='symptom_0', density=True)

plt.legend(loc='upper right')
plt.show()
```

```python
data_filtered_sym1 = df.query('(symptom_1 != 0)')
data_filtered_sym0 = df.query('(symptom_0 != 0)')

plt.hist(data_filtered_sym1['symptom_1'], bins=10, alpha=0.25, label='symptom_1')
plt.hist(data_filtered_sym0['symptom_0'], bins=10, alpha=0.25, label='symptom_0')

plt.axvline(data_filtered_sym1['symptom_1'].mean(), color='blue', linestyle='dashed', linewidth=1)
plt.axvline(data_filtered_sym0['symptom_0'].mean(), color='red', linestyle='dashed', linewidth=1)


plt.text(data_filtered_sym1['symptom_1'].mean() + 1, 450, 'Mean: {:.2f}'.format(data_filtered_sym1['symptom_1'].mean()), color='blue')
plt.text(data_filtered_sym0['symptom_0'].mean() + 1, 400, 'Mean: {:.2f}'.format(data_filtered_sym0['symptom_0'].mean()), color='red')


plt.legend(loc='upper right')
plt.show()
```

```python
plt.hist(data_filtered_sym1['symptom_1'], bins=10, alpha=0.25, label='symptom_1', density=True)
plt.hist(data_filtered_sym0['symptom_0'], bins=10, alpha=0.25, label='symptom_0', density=True)

plt.legend(loc='upper right')
plt.show()
```
